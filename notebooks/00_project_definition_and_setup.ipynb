
# Election Bloc Change Prediction Project  
## Notebook 00 — Project Definition, Experimental Design and Setup

### Project purpose

This project studies whether demographic, socioeconomic and locality-level changes can help explain and predict **changes in voting patterns between elections**, beyond a simple baseline that assumes each locality will vote similarly to the previous election.

The project is designed to address a central methodological concern:

> Locality-level voting patterns are highly stable over time. Therefore, a model can appear highly accurate even if it mainly reproduces persistent voting patterns rather than learning meaningful relationships from socioeconomic and demographic variables.

The revised project will therefore focus on:

1. Comparing all models against a strong previous-election baseline.
2. Modeling change rather than only the absolute voting level.
3. Evaluating performance both on future elections and on localities not seen during training.
4. Examining whether demographic and socioeconomic variables provide measurable added value.
5. Identifying localities whose voting behavior is unusual relative to the model's expectations.
6. Producing a scenario-based 2023 forecast and a small HTML demonstration.



## 1. Business problem

Election forecasts are usually presented at the national level, while political behavior differs greatly across localities.

However, predicting the absolute vote share of each bloc may have limited practical value because locality voting patterns are often highly stable. A simple forecast that copies the previous election result may already perform very well.

The revised business question is therefore:

> Can changes in demographic, socioeconomic and locality-level characteristics improve the prediction of changes in bloc vote shares beyond the previous-election baseline?

A secondary analytical question is:

> Which localities behave differently from what would be expected based on their previous voting pattern and observed characteristics?

### Potential value

The project may support:

- Identification of localities with meaningful political change.
- Detection of unusual or anomalous voting behavior.
- Better understanding of whether socioeconomic change is associated with political change.
- Comparison between broad national stability and within-group variation.
- Scenario-based forecasting at the locality level.



## 2. Research questions

### Primary research question

To what extent can demographic and socioeconomic variables improve the prediction of locality-level changes in voting-bloc shares relative to a persistence baseline?

### Secondary research questions

1. How strong is the previous-election baseline?
2. Does locality type alone explain most of the predictive performance?
3. Do demographic and socioeconomic variables add value after accounting for the previous election?
4. Are results different for Arab and non-Arab localities?
5. Can the model generalize to localities that were not seen during training?
6. Which localities show the largest unexplained deviations from the model?
7. Which features are most associated with predicted changes?



## 3. Unit of analysis

The main modeling table will contain one row per:

> **Locality × transition between two consecutive elections**

Example:

| locality | previous election | current election |
|---|---|---|
| Haifa | Knesset 21 | Knesset 22 |
| Haifa | Knesset 22 | Knesset 23 |
| Haifa | Knesset 23 | Knesset 24 |
| Haifa | Knesset 24 | Knesset 25 |

This structure allows the model to learn change between elections rather than only the long-term level of support in each locality.



## 4. Targets

The project models four political blocs:

- `Right_pct`
- `Center_Left_pct`
- `Haredi_pct`
- `Arab_pct`

The four shares form compositional data because they must be positive and sum to 100%.

### Main target

The preferred target is the change in CLR space:

\[
\Delta CLR = CLR(Current) - CLR(Previous)
\]

The predicted current composition will be reconstructed as:

\[
Predicted\ CLR(Current) = CLR(Previous) + Predicted\ \Delta CLR
\]

The inverse CLR transformation will return valid positive vote shares that sum to 100%.

### Additional reporting target

For interpretation, the project will also report percentage-point changes:

\[
\Delta Bloc_{pct} = Bloc_{pct,current} - Bloc_{pct,previous}
\]



## 5. Baselines

A model will only be considered useful if it is compared against strong and transparent baselines.

### Baseline A — Previous election / persistence baseline

\[
Predicted(Current) = Previous
\]

Equivalent delta prediction:

\[
Predicted\ \Delta = 0
\]

This is the main baseline.

### Baseline B — Mean change in the training data

Each bloc receives the average historical change observed in the training period.

### Baseline C — Mean change by locality group

The expected change is estimated separately by locality type or population group.

### Baseline D — Locality-type-only model

A model that uses only locality-type variables will test how much of the result is explained by broad group membership.

### Added-value tests

The main comparisons will be:

1. Previous election only.
2. Locality type only.
3. Socioeconomic and demographic features only.
4. Previous election + static locality characteristics.
5. Previous election + demographic and socioeconomic changes.



## 6. Evaluation design

The project will use more than one evaluation setting.

### A. Temporal generalization

Purpose:

> Can the model predict a later election using earlier elections?

The final election must remain outside the model-selection process.

### B. Unseen-locality generalization

Purpose:

> Can the model generalize to localities that did not appear in training?

All records from the same locality must remain in the same split.

`locality_symbol` will be used as a grouping variable, but never as a predictive feature.

### C. Group-level evaluation

Metrics will be reported separately for:

- Arab / non-Jewish localities.
- Non-Arab localities.
- Cities.
- Moshavim.
- Kibbutzim.
- Other locality types.
- Each voting bloc.

### D. Forecast evaluation metrics

Main metrics:

- MAE of final vote shares.
- MAE of vote-share changes.
- Vote-weighted MAE.
- Improvement over the previous-election baseline.
- R², reported carefully and separately by relevant subgroup.
- Bootstrap confidence intervals where feasible.



## 7. Data leakage policy

To preserve a realistic forecasting design:

- Chronological order will be respected.
- Feature filtering will be learned on training data only.
- Missing-value imputation will be fitted on training data only.
- Feature selection will be performed on training data only.
- Hyperparameter tuning will not use the final test election.
- Locality identifiers will not be used as features.
- Data from a future election will not be used to construct predictors for an earlier election.
- Any forward-fill operation will use past information only.
- The final test set will be evaluated once after the model architecture is selected.

## 9. 2023 forecast and demo

After the evaluation phase is complete, the selected architecture will be retrained using all available historical election transitions.

The 2023 forecast will use:

- Knesset 25 results as the previous-election reference.
- 2022 locality characteristics.
- 2023 locality characteristics.
- Changes between 2022 and 2023.
- The final trained preprocessing and modeling pipeline.

Because no Knesset election took place in 2023, this output will be presented as a:

> **Scenario-based forecast using 2023 locality conditions**

The project will also export a small standalone HTML demo that allows a user to:

- Select a locality.
- Compare Knesset 25 results with the 2023 scenario forecast.
- View predicted changes by bloc.
- Compare the model with the persistence baseline.
- View the main explanatory features.
- Read the methodological limitations.



## 10. Success criteria

The project will be considered successful if it demonstrates at least one of the following:

1. The model improves meaningfully over the previous-election baseline.
2. The model improves prediction within a specific subgroup.
3. The model identifies stable and interpretable predictors of change.
4. The model provides useful anomaly detection even when forecasting improvement is limited.
5. The model generalizes to unseen localities better than a simple locality-type baseline.

A high overall R² alone will not be treated as sufficient evidence of success.



## 11. Limitations defined in advance

The project does not attempt to:

- Predict individual voting behavior.
- Predict national seat allocation.
- Model campaign events, candidate effects or security shocks directly.
- Claim causal relationships between demographic variables and voting.
- Treat a 2023 scenario forecast as an observed-election prediction.
- Assume that performance on known localities automatically implies performance on new localities.

These limitations will be reported explicitly in the final presentation.
